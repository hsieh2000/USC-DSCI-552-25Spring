#### i

- Minimum: Represents the minimum time series data within a window.  
- Maximum: Represents the maximum time series data within a window.  
- Mean: The average of time series data over a window.  
- Median: The middle value once time series data are arranged in order, which is also the second quartile.  
- First quartile - The value of 25th percentile of time series data.
- Third quartile - The value of 75th percentile of time series data.
- Interquartile Range - The gap between 75th and 25th percentiles of time series data.
- Variance: The dispersion of numbers from the mean.  
- Standard deviation: The square root of variance.  
- Range: The difference between the maximum and minimum.  
- Skewness: The measure of symmetry, or more precisely, the lack of symmetry.  
- Kurtosis: The measure of whether the data are heavy-tailed or light-tailed relative to a normal distribution.

```python

df_confidence_intervals = pd.DataFrame()
for col in df_tdf.columns:

    data = df_tdf[col].to_numpy()
    # print(data)
    data = data.reshape(1,-1)
    # print(data.shape)
    
    res = bootstrap(data, np.std, confidence_level = 0.9, random_state = 42, method = 'percentile')
    _df = pd.DataFrame({'time-domain features': col, 'lower bound': res.confidence_interval.low, 'upper bound': res.confidence_interval.high}, index=[0])
    df_confidence_intervals = pd.concat([df_confidence_intervals, _df], ignore_index =True)

df_confidence_intervals
```

We can expect that the training residual sum of squares (RSS) for the cubic regression would be lower than the RSS for linear regression, because even the relationship between predictor and quantitative response is linearity, there still some noises or outliers exist.
By applying cubic regression, since cubic regression is more flexible and has more parameters, we are able to fit these noises and outliers. However it will cause overfitting, so the training RSS might be lower, but the testing RSS will spike.

#### (b)
Just like what I mentioned in (a), the testing RSS will for the cubic regression will increase significantly than the linear regression, because the cubic regression wrongly learing all noises and outliers from the sample data, sample data might not have the same distribution with the population, that's why the testing RSS will spike. So, the training RSS for linear regression will be lower.

#### (c)  
The training RSS for the cubic regression in this case might be lower.  
In general, cubic regression has better flexibility to fit more complex patterns. If the relationship is non-linear, even if the cubic regression can not capture the pattern well, it still performs better than the linear regression. If the relationship is more like linear, then the cubic regression and the linear regression may perform similarly.
